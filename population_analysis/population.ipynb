Импортируем библиотеки pandas, seaborn, matplotlib 

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

Открываем файл и делаем первичный анализ

```python
wp = pd.read_csv('world_population.csv')
wp.head()
```

Информация о данных

```python
wp.info()
```

Пустые значения

```python
wp.isnull().sum()
```

Как мы видим, пустые значения отсутсвуют

Затем подготавливаем данные для дальнейшей работы с ними, удаляем ненужные колонки

```python
wp.drop(['2015 Population', '2010 Population', '1990 Population', '1980 Population', 'CCA3', 'Rank'], axis= 1 , inplace= True)
```

1. Какие страны входят в ТОП-10 по населению в 2022 году?

```python
top_10_2022 = wp.nlargest(10, '2022 Population')
top_10_2022
```

Составляем диаграмму на основе полученного результата

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=top_10_2022['Country/Territory'], y=top_10_2022['2022 Population'], marker='.')
plt.xlabel('Страна')
plt.ylabel('Население (млрд человек)')
plt.title('ТОП-10 стран по населению')
plt.xticks(rotation=10)
plt.show()
```

2. Какие страны в ТОП-10 по населению в 1970 году?

```python
top_10_1970 = wp.nlargest(10, '1970 Population')
top_10_1970
```

Сравниваем результаты, и делаем вывод о том, что ТОП-10 стран по населению изменился за последние 50 лет. Вместо Японии и Германии в 1970, в ТОП-10 в 2022 году попадают Нигерия и Мексика. Также Россия и США опустились на несколько позиций ниже, а Пакистан и Бразилия наоборот поднялись повыше. Первые 2 места за собой сохранили Китай и Индия, сохранив большее представительство азиатских стран в ТОП-10.

3. Какие страны входят в ТОП-10 стран по площади? 

```python
top_10_size = wp.nlargest(10, 'Area (km²)')
top_10_size
```

Составляем столбчатую диаграмму на основе полученных данных

```python
plt.figure(figsize=(20,10))
sns.barplot(x=top_10_size['Country/Territory'], y=top_10_size['Area (km²)'])
plt.xticks(rotation=10)
plt.xlabel('Страна')
plt.ylabel('Площадь')
plt.show()
```

4. Каких стран больше: с приростом или убылью населения?

Добавляем столбец с данными о том, прирост населения в стране или убыль

```python
check_grow = []
for value in wp['Growth Rate']:
    if value > 1:
        check_grow.append('up')
    elif value < 1:
        check_grow.append('down')
    else:
        check_grow.append('stay')
        
wp['Check grow'] = check_grow
wp
```

Составляем круглую диаграмму на основе полученных данных

```python

plt.figure(figsize=(10, 10))
plt.pie(wp['Check grow'].value_counts(), labels= ['Прирост', 'Убыль', 'Стагнация'], autopct='%1.1f%%')
plt.title('Соотношение стран с приростом и убылью населения')
plt.show()
```

Делаем вывод о том, что в мире преобладает тенденции роста населения в большем количестве стран, однако во многих стран наблюдается убыль населения

6. Во сколько раз выросло население в самой маленькой по площади стране Азии с 1970 года по 2020 год?

```python
#фильтруем азиатские страны
asia = wp[wp['Continent'] == 'Asia']
asia.head()
```
